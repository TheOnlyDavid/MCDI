### Introducción:

En el ámbito de la ciencia de la computación, entender cómo crece el tiempo de ejecución de un algoritmo a medida que el tamaño de la entrada aumenta es fundamental para evaluar su eficiencia. Los órdenes de crecimiento son una forma de describir cómo se incrementa el tiempo de ejecución de un algoritmo en relación con el tamaño de la entrada. En este ejercicio, exploramos cinco comparaciones de órdenes de crecimiento comunes y generamos visualizaciones para entender mejor estas diferencias.

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(1, 1000, 100)  # Rango de valores de n

# Funciones correspondientes a los órdenes de crecimiento
O_1 = np.ones_like(n)
O_logn = np.log2(n)

plt.plot(n, O_1, label='O(1)')
plt.plot(n, O_logn, label='O(log n)')
plt.xlabel('n')
plt.ylabel('Tiempo de ejecución')
plt.title('Comparación: O(1) vs O(log n)')
plt.legend()
plt.show()
```

La primera gráfica muestra la comparación entre el orden de crecimiento constante (O(1)) y el orden de crecimiento logarítmico (O(log n)). En esta comparación, se puede observar que el tiempo de ejecución para el orden de crecimiento constante se mantiene constante independientemente del tamaño de la entrada (representado por "n"). Por otro lado, el orden de crecimiento logarítmico muestra un aumento gradual del tiempo de ejecución a medida que "n" aumenta, pero a un ritmo mucho más lento en comparación con los órdenes de crecimiento superiores.

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(1, 1000, 100)  # Rango de valores de n

# Funciones correspondientes a los órdenes de crecimiento
O_n = n
O_nlogn = n * np.log2(n)

plt.plot(n, O_n, label='O(n)')
plt.plot(n, O_nlogn, label='O(n log n)')
plt.xlabel('n')
plt.ylabel('Tiempo de ejecución')
plt.title('Comparación: O(n) vs O(n log n)')
plt.legend()
plt.show()
```

En la segunda gráfica, se compara el orden de crecimiento lineal (O(n)) con el orden de crecimiento lineal multiplicado por el logaritmo del tamaño de la entrada (O(n log n)). Aquí se puede observar que el tiempo de ejecución para el orden de crecimiento lineal aumenta de manera proporcional al tamaño de la entrada. Por otro lado, el orden de crecimiento O(n log n) muestra un aumento más rápido a medida que "n" aumenta, debido a la multiplicación por el logaritmo.

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(1, 100, 100)  # Rango de valores de n

# Funciones correspondientes a los órdenes de crecimiento
O_n2 = n ** 2
O_n3 = n ** 3

plt.plot(n, O_n2, label='O(n^2)')
plt.plot(n, O_n3, label='O(n^3)')
plt.xlabel('n')
plt.ylabel('Tiempo de ejecución')
plt.title('Comparación: O(n^2) vs O(n^3)')
plt.legend()
plt.show()
```

En la tercera gráfica, se comparan dos órdenes de crecimiento polinomiales: O(n^2) y O(n^3). En esta comparación, se puede observar que el tiempo de ejecución para el orden de crecimiento O(n^2) aumenta de forma cuadrática a medida que el tamaño de la entrada aumenta. Por otro lado, el orden de crecimiento O(n^3) muestra un aumento aún más rápido, ya que el tiempo de ejecución aumenta de manera cúbica con el tamaño de la entrada.

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(0, 10, 100)  # Rango de valores de n

# Funciones correspondientes a los órdenes de crecimiento
a_n = 2 ** n
n_factorial = [np.math.factorial(int(x)) for x in n]

plt.plot(n, a_n, label='a^n')
plt.plot(n, n_factorial, label='n!')
plt.xlabel('n')
plt.ylabel('Tiempo de ejecución')
plt.title('Comparación: a^n vs n!')
plt.legend()
plt.show()
```

La cuarta gráfica compara el orden de crecimiento exponencial (a^n) con el orden de crecimiento factorial (n!). Aquí se puede observar que el tiempo de ejecución para el orden de crecimiento exponencial aumenta de manera drástica a medida que "n" aumenta, mostrando un crecimiento extremadamente rápido. En contraste, el orden de crecimiento factorial también aumenta rápidamente, pero a un ritmo ligeramente más lento en comparación con el orden de crecimiento exponencial.

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(1, 10, 10)  # Rango de valores de n

# Funciones correspondientes a los órdenes de crecimiento
n_factorial = [np.math.factorial(int(x)) for x in n]
n_to_the_n = n ** n

plt.plot(n, n_factorial, label='n!')
plt.plot(n, n_to_the_n, label='n^n')
plt.xlabel('n')
plt.ylabel('Tiempo de ejecución')
plt.title('Comparación: n! vs n^n')
plt.legend()
plt.show()
```

La última gráfica muestra la comparación entre el orden de crecimiento factorial (n!) y el orden de crecimiento exponencial (n^n). Aquí se puede observar que el tiempo de ejecución para el orden de crecimiento factorial aumenta rápidamente a medida que el tamaño de la entrada aumenta, pero a un ritmo más lento en comparación con el orden de crecimiento exponencial. El orden de crecimiento exponencial muestra un aumento extremadamente rápido, superando al orden de crecimiento factorial a medida que "n" aumenta.

```python
import pandas as pd

n_values = [100, 1000, 10000, 100000]

data = {
    'n': n_values,
    'O(1)': [1 for _ in n_values],
    'O(log n)': [np.log2(n) for n in n_values],
    'O(n)': n_values,
    'O(n log n)': [n * np.log2(n) for n in n_values],
    'O(n^2)': [n ** 2 for n in n_values],
    'O(n^3)': [n ** 3 for n in n_values],
    'a^n': [2 ** n for n in n_values],
    'n!': [np.math.factorial(n) for n in n_values],
    'n^n': [n ** n for n in n_values]
}

df = pd.DataFrame(data)
```
