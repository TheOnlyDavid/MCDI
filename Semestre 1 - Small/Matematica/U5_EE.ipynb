### A

Construye un modelo lineal para los datos implementando en Python la fórmula 5.58 de la lectura, utilizando Numpy para realizar los cálculos. 

Para encontrar el modelo de regresión lineal, utilizamos la fórmula de regresión lineal $( p_1(t) = \alpha + \beta t )$ para originar la ecuación 5.58 mostrada en el libro, lo cual se conoce como aproximación por una función lineal

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math, Latex
```

```python
# Datos de la tabla
t = np.array([1, 2, 3, 4, 5])
y = np.array([55.2, 44.9, 37.9, 35.3, 30.1])
```

```python
# Calcular las entradas de la ecuación 5.58
n = len(t)
sum_t = np.sum(t)
sum_y = np.sum(y)
sum_t2 = np.sum(t**2)
sum_ty = np.sum(t*y)
```

```python
# Resolver la ecuación lineal 5.58
a = np.array([[n, sum_t], [sum_t, sum_t2]])
b = np.array([sum_y, sum_ty])
res = np.linalg.solve(a, b)
```

```python
print("Modelo de regresión lineal: p_1(t) = {:.2f} + {:.2f}t".format(res[1], res[0]))
```

Entonces tenemos $\alpha = -5.98 ,\beta = 58.62$

```python
plt.scatter(t, y)
plt.plot(t, res[0] + res[1] * t, color='red')
plt.xlabel("Años de uso")
plt.ylabel("Precio (miles de euros)")
plt.title("Precio BMW (Modelo lineal)")
plt.show()
```

La gráfica resultante muestra los puntos de datos de la tabla y la línea de regresión. Podemos ver que la línea de regresión se ajusta bastante bien a los datos, lo que sugiere que nuestro modelo lineal es un buen ajuste para estos.

### B

Una aproximación cuadrática para la forma $p_2(t) = \alpha + \beta t +\gamma t^2$ se conoce como aproximación por función cuadrática y sirve para originar la fórmula 5.65 del libro

NumPy tiene una función llamada polyfit que se puede utilizar para ajustar una curva polinómica a un conjunto de datos, incluyendo una curva cuadrática. La función polyfit ajusta una curva polinómica de grado "n" a los datos, devolviendo los coeficientes del polinomio en orden descendente.

En este caso, queremos ajustar una curva cuadrática, por lo que "n" será igual a 2. Podemos usar la función polyfit de la siguiente manera:

```python
# Encontrar los coeficientes de la curva cuadrática
coef = np.polyfit(t, y, 2)
```

```python
print("Modelo cuadrático: p_2(t) = {:.2f} {:.2f}t + {:.2f}t^2".format(coef[2], coef[1], coef[0]))
```

Entonces tenemos $\alpha = 65.92 ,\beta = -12.24, \gamma = 1.04 $

También, usando las entradas de la fórmula 5.65 y la función de Numpy linalg.solve, se tiene:

```python
# Calcular las entradas de la ecuación 5.65
n = len(t)
sum_t = np.sum(t)
sum_y = np.sum(y)
sum_t2 = np.sum(t**2)
sum_t3 = np.sum(t**3)
sum_t4 = np.sum(t**4)
sum_ty = np.sum(t*y)
sum_yt2 = np.sum(y*(t**2))

# Resolver la ecuación lineal 5.65
a = np.array([[n, sum_t, sum_t2], [sum_t, sum_t2, sum_t3], [sum_t2, sum_t3, sum_t4]])
b = np.array([sum_y, sum_ty,sum_yt2])
res = np.linalg.solve(a, b)
print("Modelo cuadrático: p_2(t) = {:.2f} {:.2f}t + {:.2f}t^2".format(res[0], res[1], res[2]))
```

```python
plt.scatter(t, y)

# Graficar la curva cuadrática
x_values = np.linspace(0, 6)
y_values = 1.04*x_values**2 - 12.24*x_values + 65.92
plt.plot(x_values, y_values, color="red")

plt.xlabel("Años de uso")
plt.ylabel("Precio (miles de euros)")
plt.title("Precio BMW (Modelo cuadrático)")
plt.show()
```

Nuevamente nos percatamos de manera visual que nuestro modelo tiene un comportamiento bueno respecto a los datos de entrada, esto para el caso del modelo cuadrático.
