```python
from mpi4py import MPI

def sieve_of_eratosthenes(n, rank, size):
    # Calculate the chunk size for each process
    chunk_size = n // size
    start = rank * chunk_size + 2  # Start from 2 to n
    end = (rank + 1) * chunk_size + 2 if rank != size - 1 else n

    # Create a boolean array "primes[0..n]" and initialize all entries as true.
    # A value in primes[i] will finally be false if i is not a prime, else true.
    primes = [True] * (end - start)

    p = 2
    while p * p <= n:
        # If primes[p] is not changed, then it is a prime
        if primes[p - start]:
            # Update all multiples of p within the process's range
            for i in range(max(p * p, (start + p - 1) // p * p), end, p):
                primes[i - start] = False
        p += 1

    return [p for p, is_prime in enumerate(primes, start) if is_prime]

if __name__ == '__main__':
    comm = MPI.COMM_WORLD
    rank = comm.Get_rank()
    size = comm.Get_size()

    n = 100  # Rango de números para encontrar primos

    # Scatter the work among processes
    local_primes = sieve_of_eratosthenes(n, rank, size)

    # Gather the results from all processes
    all_primes = comm.gather(local_primes, root=0)

    if rank == 0:
        # Combine the lists from all processes to get the final list of primes
        prime_numbers = [num for sublist in all_primes for num in sublist]
        print("Números primos en el rango de 2 a", n, ":", prime_numbers)
```
